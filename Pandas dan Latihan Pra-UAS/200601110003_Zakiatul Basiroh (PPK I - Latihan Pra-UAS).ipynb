### 1. Pola Karakter
#### Deskripsi
Pak Dono memiliki hobi menggambar pola unik. Ia memiliki sebuah kertas yang dibagi menjadi petak-petak sebanyak N baris dan M kolom. Baris-baris dinomori dari 1 sampai dengan N sedangkan kolom-kolom dinomori dari 1 sampai dengan M. Petak (a, b) menyatakan petak yang berada pada baris ke-a dan kolom ke-b. Kemudian ia menggambar simbol-simbol pada kertas tersebut dengan pola unik berikut.
- Petak (a, b) diberi simbol '#', jika a dan b keduanya genap,
- Petak (a, b) diberi simbol '*', jika a dan b keduanya ganjil,
- Petak (a, b) diberi simbol '$', jika hanya salah satu di antara a dan b yang genap.

Sayangnya, setelah menggambar pola unik, kertas gambar tersebut hilang.
Bantulah Pak Dono untuk menggambar kembali pola unik yang ia buat. 

#### Format Masukan
Sebuah baris berisi dua buah bilangan bulat N dan M dipisahkan oleh sebuah spasi.
#### Format Keluaran
N buah baris masing-masing berisi M buah karakter yang menyatakan simbol yang Pak Dono gambar.
#### Batasan
- 1 ≤ N ≤ 100
- 1 ≤ M ≤ 100

```python
Baris, Kolom = map(int, input().split())
for a in range (1, Baris + 1):
    for b in range (1, Kolom + 1):
        if a % 2 == 1 or a == 1:
            if b % 2 == 1:
                print("*", end = "")
            else:
                print("$", end = "")
        else:
            if b % 2 == 1:
                print("$", end = "")
            else:
                print("#", end = "")
    print()
```
