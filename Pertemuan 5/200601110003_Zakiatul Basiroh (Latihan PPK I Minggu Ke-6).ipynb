#### No. 1
Buatlah program yang meminta masukan user sebuah bilangan bulat N dimana (N > 0). Program kemudian menampilkan penjumlahan N bilangan genap positif pertama (bilangan genap >= 0).

```python
print('Menghitung Jumlah N Bilangan Genap Positif Pertama')
print(' ')

def pengulangan():
    Bilangan_Bulat = int(input('Masukkan Nilai N : '))
    count = 0
    sum = 0
    for i in range (0,Bilangan_Bulat*2,2):
        if i % 2 == 0:
            count = count + 1
            sum = sum + i
            print('Bilangan Genap Positif Ke -',count, ' :',i)
    print()       
    print('Jika user memasukkan N = ',Bilangan_Bulat,', maka hasil penjumlahan: ',sum)
    
    print ('')
    print ('Ingin Memasukkan Nilai Lagi? (Ya/Tidak)')
    x = input()
    if x == 'Ya':
        return pengulangan()
    else:
        print('Stop')
        
pengulangan()
```

#### No. 2
Buatlah sebuah program yang meminta masukan user sebuah bilangan bulat N dimana (N > 0). Kemudian, program menampilkan penjumlahan N bilangan kuadrat pertama. Bilangan kuadrat adalah = 1, 4, 9, 16, 25, 36, ...., N^2.

```python
print('Menghitung Jumlah N Bilangan Kuadrat Pertama')
print(' ')

def pengulangan():
    Bilangan_Bulat = int(input('Masukkan Nilai N : '))
    count = 0
    sum = 0
    for i in range(1,Bilangan_Bulat+1):
        if i > 0:
            count = count + 1
            sum = sum + i*i
            print('Bilangan Kuadrat Ke -',count, ' : ',i*i)
    print()       
    print('Jika user memasukkan N = ',Bilangan_Bulat,', maka hasil penjumlahan kuadrat: ',sum)
    
    print ('')
    print ('Ingin Memasukkan Nilai Lagi? (Ya/Tidak)')
    x = input()
    if x == 'Ya':
        return pengulangan()
    else:
        print('Stop')
        
pengulangan()
```

Bagaimana cara menampilkan judul (title) dari grafik?

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,2,0.05)*np.pi
y = np.sin(x)

plt.figure()
ax = plt.axes()
ax.plot(x, y)
plt.title("Grafik Garis")

plt.show()
```

#### No. 2
Bagaimana cara memberikan label pada sumbu x dan y?

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,2,0.05)*np.pi
y = np.sin(x)
plt.plot(x,y, label=r'$ \mathcal{x = sumbu(x)} $')

y2 = np.cos(x)
plt.plot(x,y2,'g', label=r'$ \mathcal{y = sumbu(y)} $')

plt.legend(loc='upper right')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.0, 3.0, 0.01)
y = np.sin(2.5 * np.pi * x)
plt.plot(x, y, label=r'$ \mathcal{x = sumbu(x)} $')

y2 = np.cos(2.5 * np.pi * x)
plt.plot(x,y2,'g', label=r'$ \mathcal{y = sumbu(y)} $')
plt.legend(loc='upper right')

plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')

plt.title('Grafik Garis')

plt.show()
```
